```python
# STEP 1: Install and Import dependencies
!pip install kaggle tensorflow keras matplotlib pillow split-folders

import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing import image
from google.colab import files
import splitfolders

# STEP 2: Upload kaggle.json to Colab (from Kaggle -> Account -> API -> Create New Token)
files.upload()   # <-- Upload kaggle.json here

# STEP 3: Configure Kaggle API
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

# STEP 4: Download & Unzip Dataset
!kaggle datasets download -d muhammad0subhan/fruit-and-vegetable-disease-healthy-vs-rotten
!unzip -oq fruit-and-vegetable-disease-healthy-vs-rotten.zip -d dataset

# Check dataset contents
!ls "dataset/Fruit And Vegetable Diseases Dataset" | head -20

# STEP 5: Split dataset into Train/Val
input_folder = "dataset/Fruit And Vegetable Diseases Dataset"
output_folder = "dataset/splitted"

splitfolders.ratio(input_folder, output=output_folder, seed=42, ratio=(.8, .2))

train_dir = os.path.join(output_folder, "train")
test_dir = os.path.join(output_folder, "val")

print("✅ Fixed Paths:")
print("Train:", train_dir)
print("Test:", test_dir)

# STEP 6: Data Generators
img_size = (128, 128)
batch_size = 32

train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

val_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

test_datagen = ImageDataGenerator(rescale=1./255)
test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)

# STEP 7: Build CNN Model
num_classes = len(train_gen.class_indices)

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# STEP 8: Train Model
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10
)

# STEP 9: Evaluate
loss, acc = model.evaluate(test_gen)
print(f"Test Accuracy: {acc*100:.2f}%")

# STEP 10: Plot Accuracy & Loss
plt.plot(history.history['accuracy'], label="Train Acc")
plt.plot(history.history['val_accuracy'], label="Val Acc")
plt.legend()
plt.show()

# STEP 11: Predict on Uploaded Image
def predict_image(img_path):
    img = image.load_img(img_path, target_size=img_size)
    img_array = image.img_to_array(img)/255.0
    img_array = np.expand_dims(img_array, axis=0)
    prediction = model.predict(img_array)
    class_index = np.argmax(prediction)
    labels = list(train_gen.class_indices.keys())
    return labels[class_index]

# Upload image and test
uploaded = files.upload()
for fn in uploaded.keys():
    result = predict_image(fn)
    print(f"Prediction: {result}")
```
